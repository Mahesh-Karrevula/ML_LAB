## <a name='reshape'></a>5Ô∏è‚É£ Reshaping, Transpose & Copies vs‚ÄØViews

```python
b = np.arange(8)
B = b.reshape(2,4)
B[0,0] = 99
print('b is modified:', b)
C = b.reshape(2,4).copy()
C[0,0] = -1
print('b unchanged with copy:', b)
```

```python
# üñäÔ∏è TODO: Flatten a 3‚ÄëD array into 1‚ÄëD using both `ravel` and `flatten`; observe copy vs view.
arr3d = np.arange(1, 13).reshape(2, 2, 3)
print("Original 3D array:\n", arr3d)

r = arr3d.ravel()
print("\nUsing ravel():", r)

f = arr3d.flatten()
print("Using flatten():", f)

r[0] = 999
print("\nAfter modifying ravel() output:")
print("ravel view:", r)
print("Original array after ravel change:\n", arr3d)

f[1] = 555
print("\nAfter modifying flatten() output:")
print("flatten copy:", f)
print("Original array after flatten change:\n", arr3d)
```

```python
# üñäÔ∏è TODO: compute eigenvalues of `A` using `np.linalg.eig`.
import numpy as np

A = np.random.random((3, 3))
b = np.random.random(3)
x = np.linalg.solve(A, b)
print('A¬∑x ‚âà b?', np.allclose(A.dot(x), b))

eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues of A:", eigenvalues)

```

```python
# üñäÔ∏è TODO: replace NaNs with column means in a 2‚ÄëD array containing NaNs.
import numpy as np

arr2d = np.array([
    [1,   2,  np.nan],
    [4, np.nan, 6],
    [7,   8,  9]
], dtype=float)

col_means = np.nanmean(arr2d, axis=0)
inds = np.where(np.isnan(arr2d))
arr2d[inds] = np.take(col_means, inds[1])

print("Column means:", col_means)
print("Array after replacing NaNs:\n", arr2d)
```

## <a name='project'></a>1Ô∏è‚É£6Ô∏è‚É£ Mini‚ÄëProject: Fitness Data Analysis

Load `fitness.txt` (tab‚Äëseparated) then follow prompts.

```python
fitness = np.genfromtxt('fitness.txt', delimiter='\t', dtype=None, encoding=None, names=True)
print('columns:', fitness.dtype.names, 'rows:', len(fitness))
```

```python
# üñäÔ∏è TODO: Monthly step count, sleep vs mood correlation, weekly summary, etc.
import pandas as pd
import numpy as np

col_names = ['date', 'step_count', 'mood', 'calories_burned', 'hours_of_sleep', 'bool_of_active', 'weight_kg']

fitness = pd.read_csv('fitness.txt', sep='\s+', comment='#', names=col_names)

fitness['date'] = pd.to_datetime(fitness['date'], format='%d-%m-%Y')

dates = fitness['date'].to_numpy()
step_counts = fitness['step_count'].to_numpy()
months = fitness['date'].dt.strftime('%Y-%m').to_numpy()
unique_months = pd.unique(months)

monthly_steps = {month: step_counts[months == month].sum() for month in unique_months}

print("Monthly Step Count:")
for month, total in monthly_steps.items():
    print(f"{month}: {total}")

sleep = fitness['hours_of_sleep'].astype(float).to_numpy()
mood = fitness['mood'].astype(float).to_numpy()

correlation = np.corrcoef(sleep, mood)[0, 1]
print(f"\nCorrelation between sleep and mood: {correlation:.2f}")

weeks = fitness['date'].dt.strftime('%Y-W%U').to_numpy()
unique_weeks = pd.unique(weeks)

weekly_summary = {}
for week in unique_weeks:
    mask = weeks == week
    avg_steps = step_counts[mask].mean()
    avg_mood = mood[mask].mean()
    avg_sleep = sleep[mask].mean()
    weekly_summary[week] = (avg_steps, avg_mood, avg_sleep)

print("\n Weekly Summary (Steps, Mood, Sleep):")
for week, (steps, mood_val, sleep_val) in list(weekly_summary.items())[:5]:
    print(f"{week}: Steps={steps:.1f}, Mood={mood_val:.1f}, Sleep={sleep_val:.1f}")
```
